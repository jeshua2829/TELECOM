```python
import pandas as pd
import requests
```

```python
# URL de la API que devuelve un JSON
url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json"

# Hacemos la petici贸n GET
response = requests.get(url)

# Verificamos que la respuesta fue exitosa (c贸digo 200)
data = response.json()

Datos = pd.json_normalize(data)

Datos
```

# Transformaci贸n

```python
Datos.info()
```

```python
Datos.columns = [
    # Si es una de las columnas espec铆ficas, reemplazamos solo 'account.' por ''
    col.replace("account.", "") if col.startswith("account.Charges.") else
    # Si no es una de esas, quitamos todo el prefijo antes del 煤ltimo punto
    col.split('.')[-1]
    for col in Datos.columns
]
```
